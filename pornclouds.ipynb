```python
## TOP VIDEOS
```

```python
top = pd.read_csv('parsed.csv')
top.iloc[:, [0]]
```

```python
class Top(object):
    def __init__(self):
        self.my_txt = ''
        self.words = []

    def cut_words(self):
        with open('parsed.csv') as f:
            self.my_txt = f.read()
        words_list = jieba.lcut(self.my_txt, cut_all=False)
        self.words = [word for word in words_list if len(word) > 1]
        self.generate_wordcloud()

    def generate_wordcloud(self):
        fig = plt.figure(1, figsize=(12,30))
        word_cloud = WordCloud(background_color='white', scale=2 )
        words_img = word_cloud.generate(' '.join(self.words))
        plt.imshow(words_img)
        plt.suptitle('Hottest and Most Viewed Videos in 2021', fontsize=20)
        fig.subplots_adjust(top=2)
        plt.axis('off')
        plt.show()


if __name__ == '__main__':
    W = Top()
    W.cut_words()
```

```python
collected = pd.read_csv('woc.csv')
```

```python
from wordcloud import WordCloud, STOPWORDS

stopwords = set(STOPWORDS)
stopwords.update(["NaN", "columns", "answers", "rows", "videos", "portrayed", "typically", "see", "mainly", "categories", "gravitate", "sites", "viewing", "search", "towards", "consume", "featured", "content", "regularly", "backgrounds"])

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        min_word_length=2,
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 30))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2)

    plt.imshow(wordcloud)
    plt.show()
```

```python
show_wordcloud(collected.iloc[:,[11]], 'How do you typically see Asian/Pacific-Islander\n women portrayed in the videos you consume?')
```

```python
show_wordcloud(collected.iloc[:,[12]], 'How do you typically see Black/African-American\n women portrayed in the videos you consume?')
```

```python
#collected.iloc[:,[13]]
show_wordcloud(collected.iloc[:,[13]], 'How do you typically see Hispanic/Latin-American\n women portrayed in the videos you consume?')
```

```python
show_wordcloud(collected.iloc[:,[10]], "Out of the content you consume, which \nethnic backgrounds are featured regularly?")
```
